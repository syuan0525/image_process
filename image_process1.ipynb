```python
from PIL import ImageFont, ImageDraw, Image
```

```python
pic = np.full((300, 500, 3), (200, 255, 200), np.uint8)
cv2.line(pic, (10,10), (200,200), (0,0,255), 2)
cv2.rectangle(pic, (10,10), (200,200), (0,255,0), 2)
cv2.rectangle(pic, (10,250), (200,280), (0,255,0), -1)
cv2.circle(pic, (300,150), 50, (255,0,0), 2)
cv2.circle(pic, (400,150), 50, (255,0,0), -1)
pic = Image.fromarray(pic)
ImageDraw.Draw(pic).text((300,50), "要寫的字", (255,0,0),ImageFont.truetype(""))
# ImageDraw.Draw(PIL圖像變數).text(文字位置, 要寫的文字, 顏色, 設定)


cv2_imshow(pic)
# cv2.line(圖像變數, 線的起點, 線的終點, 顏色, 線粗細)
# cv2.rectangle(圖像變數, 矩形左上點, 矩形右下點, 顏色, 線粗細) 粗細-1實心
# cv2.circle(圖像變數, 中心點, 半徑, 顏色, 線粗細)

```
