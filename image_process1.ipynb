```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
```

```python
pic = cv2.imread("/content/th.jpg", 1)   # -1 包含透明值
cv2_imshow(pic)
"""

本地端python
cv2.imshow("window 1", pic)
cv2.waitKey(0)

"""
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY) # 色彩空間轉換
cv2_imshow(pic)

print("高: ",pic.shape[0])
print("寬: ",pic.shape[1])
```

```python
from PIL import ImageFont, ImageDraw, Image
```

```python
pic = np.full((300, 500, 3), (200, 255, 200), np.uint8) # 高,寬,維度 BGR顏色
# pic = np.full((300, 500), 127, np.uint8) # 灰階
cv2.line(pic, (10,10), (200,200), (0,0,255), 2)
cv2.rectangle(pic, (10,10), (200,200), (0,255,0), 2)
cv2.rectangle(pic, (10,250), (200,280), (0,255,0), -1)
cv2.circle(pic, (300,150), 50, (255,0,0), 2)
cv2.circle(pic, (400,150), 50, (255,0,0), -1)
pic = Image.fromarray(pic)
ImageDraw.Draw(pic).text((250,50), "要寫的字", (255,0,0), ImageFont.truetype("kaiu.ttf",50))
# ImageDraw.Draw(PIL圖像變數).text(文字位置, 要寫的文字, 顏色, 設定)

pic = np.array(pic)
cv2_imshow(pic)
# cv2.line(圖像變數, 線的起點, 線的終點, 顏色, 線粗細)
# cv2.rectangle(圖像變數, 矩形左上點, 矩形右下點, 顏色, 線粗細) 粗細-1實心
# cv2.circle(圖像變數, 中心點, 半徑, 顏色, 線粗細)

```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time
```

```python
for i in range(100, 400 ,5):
  dp.clear_output(True)
  pic = np.full((300, 500, 3), (200, 255, 200), np.uint8)
  cv2.circle(pic, (i,150), 50, (255,0,0), -1)
  cv2_imshow(pic)
  time.sleep(0.1)
```

```python
video1 = cv2.VideoWriter("1.mp4",cv2.VideoWriter_fourcc(*"MP4V"), 60, (500,300))
for i in range(100, 400 ,1):
  pic = np.full((300, 500, 3), (200, 255, 200), np.uint8)
  cv2.circle(pic, (i,150), 50, (255,0,0), -1)
  video1.write(pic)
video1.release()
```
