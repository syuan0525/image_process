```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
```

```python
pic1 = cv2.imread("1.jpg",1)
pic2 = cv2.add(pic1, (50,50,50,0))  # 加(變亮;大於255等於255)
pic3 = cv2.subtract(pic1, (50,50,50,0))  # 減(變暗;小於0等於0)
pic4 = cv2.absdiff(pic1, (200,200,200,0)) # 減完小於0會取絕對值(負片)
cv2_imshow(pic1)
cv2_imshow(pic2)
cv2_imshow(pic3)
cv2_imshow(pic4)
```

```python
pic2 = cv2.divide(pic1, (85,85,85,0)) # 除
pic3 = cv2.multiply(pic2, (85,85,85,0))  #乘 (乘除搭配使用,極色化(4值化處理))
cv2_imshow(pic3)
```

```python
pic3 = cv2.bitwise_not(pic1)  # 負片
cv2_imshow(pic3)
```

```python
pic1 = cv2.imread("1.jpg",1)
w = 100
h = int(pic1.shape[0]/(pic1.shape[1]/w))
pic2 = cv2.resize(pic1, (w,h))        # 大小調整
cv2_imshow(pic2)
```

```python
pic1 = cv2.imread("1.jpg",1)
pic2 = cv2.flip(pic1, 1)  # 1 : 左右翻轉, 0 : 上下翻轉, -1 : 上下左右都翻轉
cv2_imshow(pic2)
```

```python
pic2 = cv2.warpAffine(pic1, cv2.getRotationMatrix2D((200,200),45,0.5), (500,500))
cv2_imshow(pic2)         # (旋轉中心,旋轉角度,縮放比例) (輸出的圖像大小(寬,高)) 
```

```python
m1 = cv2.imread("1.jpg",1)
m2 = cv2.imread("2.jpg",1)

w = 200
h = int(m1.shape[0]/(m1.shape[1]/w))
m1 = cv2.resize(m1, (w,h)) 

m2[:m1.shape[0],:m1.shape[1]] = m1
# m2[200:300,200:350] = m1[50:150,50:200]
cv2_imshow(m2)
```

```python
m1 = cv2.imread("1.jpg",1)
cv2_imshow(m1[::2,::2]) # 像素每2格取一格 縮小成1/2
```

```python
m1 = cv2.imread("1.jpg",1)
m2 = cv2.imread("2.jpg",1)

w = 200
h = int(m1.shape[0]/(m1.shape[1]/w))
m1 = cv2.resize(m1, (w,h))
m2[:m1.shape[0]:2,:m1.shape[1]:2] = m1[::2,::2]  # 圖片重疊

cv2_imshow(m2)
```
