```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
```

```python
pic1 = cv2.imread("1.jpg",1)
pic2 = cv2.add(pic1, (50,50,50,0))  # 加(變亮;大於255等於255)
pic3 = cv2.subtract(pic1, (50,50,50,0))  # 減(變暗;小於0等於0)
pic4 = cv2.absdiff(pic1, (200,200,200,0)) # 減完小於0會取絕對值(負片)
cv2_imshow(pic1)
cv2_imshow(pic2)
cv2_imshow(pic3)
cv2_imshow(pic4)
```

```python
pic2 = cv2.divide(pic1, (85,85,85,0)) # 除
pic3 = cv2.multiply(pic2, (85,85,85,0))  #乘 (乘除搭配使用,極色化(4值化處理))
cv2_imshow(pic3)
```

```python
pic3 = cv2.bitwise_not(pic1)  # 負片
cv2_imshow(pic3)
```
